```python
# reads in a list of strings, and converts them to csv with correct confirmed labels. 

def convert_to_csv(full_text_dict):

    pd_df = pd.DataFrame({'premise': [], 'hypothesis': [], 'label': []})
    label_list = full_text_dict.keys()

    hypotheses = {
        "gratitude":  "This text expresses gratitude.", 
        "frustration":  "This text expresses frustration.",
        "political": "This text is political in nature.",
        "solution focsed": "This text is focused on solutions.",
        "information or news": "This text provides information or news.",
        "fear or panic": "This text expresses fear or panic.",
        "blaming": "This text contains blaming.",
        "seeking help or advice": "This text is seeking help or advice.",
        "wildfires": "This text is about wildfires.",
        "prescribed burns": "This text is about prescribed burns",
        "fire management": "This text is about fire management",
    }

    for label in full_text_dict.keys():
        current_list = full_text_dict[label]
        label_name = label 

        for current_text in current_list:
            for hypothesis in hypotheses.keys():
                premise = current_text
                hyp = hypotheses[hypothesis]

                if label == hypothesis:
                    lab = 'entailment'
                else:
                    lab = 'contradiction'

                pd_df.loc[len(pd_df)] = {
                    'premise': premise,
                    'hypothesis': hyp,
                    'label': lab
                }

    pd_csv = pd_df.to_csv("CLAUDE_training.csv", index=False)

    return pd_csv
    
# Throws all lists into a dictionary in order to make label names easier to work with.

def convert_to_dict(full_text_list):
    full_text_dict = {
        "gratitude": full_text_list[0],
        "frustration": full_text_list[1],
        "political": full_text_list[2],
        "solution focsed": full_text_list[3],
        "information or news": full_text_list[4],
        "fear or panic": full_text_list[5],
        "blaming": full_text_list[6],
        "seeking help or advice": full_text_list[7],
        "wildfires": [],
        "prescribed burns": [],
        "fire management": []
    }

    return full_text_dict
```
