```python
# del_rows = [4,6,21,43,46,52,63,70,72]

def drop_rows(del_rows):
    df = df.drop(del_rows).reset_index(drop=True)
    return df
```

```python
df["replies"] = df["replies"].apply(lambda replies: replies[:10] if isinstance(replies, list) else replies)
```

```python
# Now I have cleaned the titles so that they only pertain to wildfires, smoke, etc. 
# I want to create a function that runs sentiment analysis for each reply to each post 
# Even though I am practicing on a small dataset, I need to ensure this function can handle large datasets.
# Therefore, I can't nest loops, which is what my initial plan was. 
# ChatGPT offered a solution that would make it run faster using VADER sentiment analysis
# I will be using VADER because it is well suited for social media texts and its runtime is fast

import nltk
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

def reply_analysis(text):
    analyzer = SentimentIntensityAnalyzer()

    return analyzer.polarity_scores(text)['compound']

df['sentiment'] = df['replies'].apply(lambda replies: [reply_analysis(reply) for reply in replies])
```

```python
# Now I want to take in the averages of the sentiment analysis for each reply

def avg_sentiment(df):

    reply_scores = []

    for i in df["sentiment"]:

        if (len(i) > 0):
            average = sum(i)/len(i)

            reply_scores.append(average)
        
        else:
            reply_scores.append(0)
        
    df['average_sent_reply_score'] = reply_scores

    return df

df = avg_sentiment(df)
df
```

```python
# Using this, I want to plot average sentiment scores compared to dates 
# I will do this in Observable, time to save data !!

data = {'Date': df["date"], 
        'Score': df["average_sent_reply_score"]}
df_final = pd.DataFrame(data)
print(df_final)

df_final.to_json('reddit_data.json', 
orient='records', date_format='iso')
```
