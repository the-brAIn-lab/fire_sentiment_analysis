```python
# function that returns the distribution of lables 

def dist_labels(data_file):
    dict_distribution = {}

    label_data = data_file.drop(columns=['text', 'date'])

    for column in label_data:
        if column not in dict_distribution:
            dict_distribution[column] = 0

        for row in label_data[column]:
            if row == 1:
                dict_distribution[column] += 1
    
    return dict_distribution
```

```python
# function that combines csv files 

def combine_file(file1,file2,file3):
    combined_file = pd.concat([file1, file2, file3], ignore_index=True)
    
    return combined_file
```

```python
# function that converts a pandas file back to csv format 

def convert_to_csv(file):

    reddit_csv_file = file.to_csv("reddit_csv_file.csv", index=False)
    
    return reddit_csv_file
```
