```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_label_data = pd.read_csv("word_cloud_3_regions.csv")
```

```python
csv_label_data
```

```python
all_text = csv_label_data['text']
```

```python
import re

for text in all_text:
    words = re.findall(r'\b\w+\b', text.lower())
    print(words)
    break
```

```python
from nltk.tokenize import word_tokenize
import string

def tokenize_text(all_text):
    full_token_list = []
    for text in all_text:
        tokens = word_tokenize(text)
        tokens_filtered = [token.lower() for token in tokens if token.isalpha()]

        for word in tokens_filtered:
            full_token_list.append(word)
            
    return full_token_list


all_words = tokenize_text(all_text)
```

```python
from wordcloud import WordCloud

full_string_words = ' '.join(all_words)

wordcloud = WordCloud(width=350, height=150, background_color='white').generate(full_string_words)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image

img = wordcloud.to_image()

img.save("wordcloud.jpg", format="JPEG", quality=50, optimize=True)
```
