```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from nltk.tokenize import word_tokenize
import string
from wordcloud import WordCloud
from PIL import Image
```

```python
la_text = pd.read_csv("los_angeles_random_labels.csv")
spokane_text = pd.read_csv("spokane_random_labels.csv")
seattle_text = pd.read_csv("seattle_random_labels.csv")
slc_text = pd.read_csv("salt_lake_city_random_labels.csv")
portland_text = pd.read_csv("portland_random_labels.csv")

la_text = la_text['text']
spokane_text = spokane_text['text']
seattle_text = seattle_text['text']
slc_text = slc_text['text']
portland_text = portland_text['text']
```

```python
def tokenize_text(all_text):
    full_token_list = []
    for text in all_text:
        tokens = word_tokenize(text)
        tokens_filtered = [token.lower() for token in tokens if token.isalpha()]

        for word in tokens_filtered:
            full_token_list.append(word)
            
    return full_token_list

def create_word_cloud(text, place):
    full_string_words = ' '.join(text)

    wordcloud = WordCloud(width=350, height=150, background_color='white').generate(full_string_words)

    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(place+ " - Word Cloud")
    plt.axis('off')
    plt.show()

    return 
```

```python
all_words = tokenized_text = tokenize_text(la_text)
create_word_cloud(all_words, "LA")
```

```python
all_words = tokenized_text = tokenize_text(spokane_text)
create_word_cloud(all_words, "Spokane")
```

```python
all_words = tokenized_text = tokenize_text(seattle_text)
create_word_cloud(all_words, "Seattle")
```

```python
all_words = tokenized_text = tokenize_text(slc_text)
create_word_cloud(all_words, "Salt Lake City")
```

```python
all_words = tokenized_text = tokenize_text(portland_text)
create_word_cloud(all_words, "Portland")
```
